## Brazil

```python
eff_bra=pd.read_csv("plot/fig_s7/sensitivity_3_bra.csv")
```

```python
eff_bra
```

```python
policies = list(eff_bra['intervention'])
temp1 = []
for intervention in policies:
    for c in ['_08', '_add', '_empty1', '_empty2']:
        if c == '_08':
            temp1.append([intervention]+eff_bra[eff_bra['intervention']==intervention][['effect'+c, 'ci'+c+'_0.95']].values.tolist()[0])
        elif c == '_add':
            temp1.append([intervention+c, None, None])
        else:
            temp1.append([intervention+c, None, None])
new1 = pd.DataFrame(temp1, columns=['intervention', 'effect', 'ci_0.95'])
new1 =new1[:-2]
```

```python
new1
```

```python
policies = list(eff_bra['intervention'])
temp2 = []
for intervention in policies:
    for c in ['_08', '_add', '_empty1', '_empty2']:
        if c == '_add':
            temp2.append([intervention+c]+eff_bra[eff_bra['intervention']==intervention][['effect'+c, 'ci'+c+'_0.95']].values.tolist()[0])
        elif c == '_08':
            temp2.append([intervention, None, None])
        else:
            temp2.append([intervention+c, None, None])
new2 = pd.DataFrame(temp2, columns=['intervention', 'effect', 'ci_0.95'])
new2 =new2[:-2]
```

```python
new2
```

```python
fig1, ax = plt.subplots(figsize=[10,11])

ax.spines['left'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.axvline(x=0,ls="-",linewidth=0.6,c="black")

ax.tick_params(labelsize=15)

ax.errorbar(x=new1["effect"][::-1], 
            y=new1["intervention"][::-1], 
            xerr=new1["ci_0.95"][::-1]/2, 
            fmt="o", 
            markersize=6,
            capsize=2, 
            elinewidth=1, 
            barsabove=True, 
            color='dimgray')

ax.errorbar(x=new2["effect"][::-1], 
            y=new2["intervention"][::-1], 
            xerr=new2["ci_0.95"][::-1]/2, 
            fmt="o", 
            markersize=6,
            capsize=2, 
            elinewidth=1, 
            alpha=0.5,
            barsabove=True, 
            color='gray')

ax.set_yticks(np.linspace(0,4*12,13)+1)
plt.xticks(size=12)
plt.xlabel('Brazil',fontsize=20)
```

```python
fig1.savefig("plot/fig_s7/sensitivity_3_bra",bbox_inches='tight',dpi=1080)
```
