```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
input= pd.read_csv("/Users/ll/Documents/!Work/2020.7 Covid19/2github/plot/fig_3/prob_all.csv",index_col='intervention')
```

```python
input=input.rename(columns={'South Korea':'South\nKorea'})
```

```python
sns.palplot(sns.color_palette("rocket_r", 100)) 
```

```python
new_rocket=sns.color_palette("rocket_r", 100)[0:70]
```

```python
plt.figure(figsize=[6.5,18])
fig=sns.heatmap(input, 
                annot=True, 
                fmt='.2f', 
                annot_kws={"fontsize":12}, 
               
                cmap=new_rocket,
                cbar=False)
plt.tick_params(labelsize=12)
fig.xaxis.set_ticks_position('top')
font = {'weight' : 'normal','size' : 20}
plt.ylabel(' ',font)
```

```python
x = fig.get_figure()
x.savefig("plot/fig_3/prob_all.png",bbox_inches='tight',dpi=1080)
```
