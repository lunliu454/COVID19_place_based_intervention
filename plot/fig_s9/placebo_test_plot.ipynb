## Brazil

```python
eff_bra=pd.read_csv("/Users/ll/Documents/!Work/2020.7 Covid19/2github/plot/fig_s9/placebo_bra.csv")
```

```python
eff_bra
```

```python
policies = list(eff_bra['policy'])
temp1 = []
for policy in policies:
    for c in ['_08', '_placebo', '_empty1', '_empty2']:
        if c == '_08':
            temp1.append([policy+c]+eff_bra[eff_bra['policy']==policy][['effect'+c, 'ci'+c+'_0.95']].values.tolist()[0])
        elif c == '_placebo':
            temp1.append([policy, None, None])
        else:
            temp1.append([policy+c, None, None])
new1 = pd.DataFrame(temp1, columns=['policy', 'effect', 'ci_0.95'])
new1 =new1[:-2]
```

```python
new1
```

```python
policies = list(eff_bra['policy'])
temp2 = []
for policy in policies:
    for c in ['_08', '_placebo', '_empty1', '_empty2']:
        if c == '_placebo':
            temp2.append([policy]+eff_bra[eff_bra['policy']==policy][['effect'+c, 'ci'+c+'_0.95']].values.tolist()[0])
        else:
            temp2.append([policy+c, None, None])
new2 = pd.DataFrame(temp2, columns=['policy', 'effect', 'ci_0.95'])
new2 =new2[:-2]
```

```python
new2
```

```python
fig1, ax = plt.subplots(figsize=[10,11])

ax.spines['left'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.axvline(x=0,ls="-",linewidth=0.6,c="black")

ax.tick_params(labelsize=15)

ax.errorbar(x=new1["effect"][::-1], 
            y=new1["policy"][::-1], 
            xerr=new1["ci_0.95"][::-1]/2, 
            fmt="o", 
            markersize=6,
            capsize=2, 
            elinewidth=1, 
            barsabove=True, 
            color='dimgray')

ax.errorbar(x=new2["effect"][::-1], 
            y=new2["policy"][::-1], 
            xerr=new2["ci_0.95"][::-1]/2, 
            fmt="o", 
            markersize=6,
            capsize=2, 
            elinewidth=1, 
            alpha=0.5,
            barsabove=True, 
            color='gray')

ax.set_yticks(np.linspace(0,4*12,13)+0)
plt.xticks(size=12)
plt.xlabel('Brazil',fontsize=20)
```

```python
fig1.savefig("/Users/zhangzhu/Desktop/placebo/placebo_bra",bbox_inches='tight',dpi=1080)
```
